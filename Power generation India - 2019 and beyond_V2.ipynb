## State wise installed capacity in India and planned future for renewables 
* Data obtained from https://data.gov.in 
* Some datasets were easier to look up with simple search terms and some were harder
* All datasets were up to date as of July 2019 
* Some dataset csvs were modified to match names and for easier mapping with the map

Goal of this exercise is to map the installed capacity as of July 2019 and map the 2022 renewable plan

```python
## Import libraries - pandas used for all data IO and storage, geopandas makes mapping easy 
import pandas as pd
import matplotlib.pyplot as plt 
import geopandas as gpd
import numpy as np
%matplotlib inline
```

```python
# Creating data mapping for State names to match GIS shapefile used in this study 
csvNames = ['Andaman and Nicobar Islands', 'Chhatisgarh', 'Jammu and Kashir', 'Lakshdweep', 'Maharashra', 'Odish', 'Uttarakhand']
mapNames = ['Andaman and Nicobar', 'Chhattisgarh', 'Jammu and Kashmir', 'Lakshadweep', 'Maharashtra', 'Orissa', 'Uttaranchal']
mappingDict = dict(zip(csvNames, mapNames))
```

```python
# Reading installed capacity as of July 2019
data = pd.read_csv('MOP_installed_capacity_sector_mode_wise.csv')
data = data.replace({'&': 'and'}, regex=True)
data = data.replace({'sh.': 'sh'}, regex=True)
data.head()
```

```python
# Mapping state names 
data['State'] = data['State'].apply(lambda x: mappingDict[x] if x in mappingDict.keys() else x)
data['State'].unique()
```

```python
# Loading map shape file 
india = gpd.read_file('IND_adm/IND_adm1.shp')
india.head()
```

```python
# Renaming 'State' column for ease of use and dissolving districts into states. Data resolution is at state level. 
india.rename(columns={'NAME_1': 'State'}, inplace=True)
india = india.dissolve(by='State', as_index=False)
india.head()
```

```python
# Geopandas makes plotting column data easy. Hence, data of interest is added as column in the geopandas datastructure 
thermalPower = data[data.Mode == 'Thermal'].groupby(by='State').sum().to_dict()['Installed Capacity']
india['thermalPower'] = india['State'].apply(lambda x: thermalPower[x] if x in thermalPower.keys() else 0)

nuclearPower = data[data.Mode == 'Nuclear'].groupby(by='State').sum().to_dict()['Installed Capacity']
india['nuclearPower'] = india['State'].apply(lambda x: nuclearPower[x] if x in nuclearPower.keys() else 0)

hydroPower = data[data.Mode == 'Hydro'].groupby(by='State').sum().to_dict()['Installed Capacity']
india['hydroPower'] = india['State'].apply(lambda x: hydroPower[x] if x in hydroPower.keys() else 0)

resPower = data[data.Mode == 'RES'].groupby(by='State').sum().to_dict()['Installed Capacity']
india['resPower'] = india['State'].apply(lambda x: resPower[x] if x in resPower.keys() else 0)
```

```python
# Mapping column names to names we would like to see on the plots 
plotNames = {'thermalPower': 'Thermal', 'nuclearPower': 'Nuclear', 'hydroPower': 'Hydro',
             'resPower':'Renewables', 'windPower': 'Wind', 'solarPower': 'Solar'}
plotNames_plan = {'biomassPlanned': 'Biomass Planned', 'smallHydroPlanned': 'Small Hydro Planned',
                  'windPlanned': 'Wind Planned', 'solarPlanned': 'Solar Planned'}
```

```python
# Function to create a 2x2 plot layout 
def createMapGrid(listOfData, listOfNames, basemap):
    maxPower = max(np.max(india[listOfData]))
    fig, ax = plt.subplots(2,2, figsize=(12, 12))
    counter = 0
    for i in ax.flat:
            india.plot(ax = i, column=listOfData[counter], cmap='RdPu', vmin=0, vmax= maxPower, linewidth=0.25, edgecolor='black', legend=False)
            i.set_axis_off()
            i.set_title(listOfNames[listOfData[counter]])
            counter +=1

    ax_cbar = fig.add_axes([0.95,0.3,0.01,0.4])
    ax_cbar.set_title('Installed Capacity [MW]')
    mappable = ax[1, 1].collections[0]
    cbar = fig.colorbar(cax = ax_cbar, mappable=mappable)
    fig.subplots_adjust(hspace=0.2, wspace=-0.3)
```

```python
createMapGrid(['thermalPower', 'nuclearPower', 'hydroPower', 'resPower'], plotNames, india)
```

```python
windInstalled = pd.read_csv('state_wise_wind.csv', index_col='State').to_dict()['Installed Capacity']
india['windPower'] = india['State'].apply(lambda x: windInstalled[x] if x in windInstalled.keys() else 0)

solarInstalled = pd.read_csv('installed_solar_capacity.csv', index_col='State').to_dict()['Installed Capacity']
india['solarPower'] = india['State'].apply(lambda x: solarInstalled[x] if x in solarInstalled.keys() else 0)
```

```python
createMapGrid(['nuclearPower', 'hydroPower', 'windPower','solarPower'], plotNames, india)
```

```python
targets_2022 = pd.read_csv('renewable_targets_2022.csv', index_col = 'State').to_dict()
solarTarget = targets_2022['Solar']
india['solarPlanned'] = india['State'].apply(lambda x: solarTarget[x] if x in solarTarget.keys() else 0)

windTarget = targets_2022['Wind']
india['windPlanned'] = india['State'].apply(lambda x: windTarget[x] if x in windTarget.keys() else 0)

shpTarget = targets_2022['SHP']
india['smallHydroPlanned'] = india['State'].apply(lambda x: shpTarget[x] if x in shpTarget.keys() else 0)

bioTarget = targets_2022['Biomass']
india['biomassPlanned'] = india['State'].apply(lambda x: bioTarget[x] if x in bioTarget.keys() else 0)
```

```python
createMapGrid(['solarPlanned', 'windPlanned', 'smallHydroPlanned', 'biomassPlanned'], plotNames_plan, india)
```
